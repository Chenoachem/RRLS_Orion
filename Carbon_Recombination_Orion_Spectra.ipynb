```python
import os
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import sys
import csv
from operator import itemgetter, attrgetter, methodcaller
%matplotlib inline
```

```python
data = np.genfromtxt('HD33142.csv', delimiter=',', skip_header=10,skip_footer=10, names=['x', 'y'])
```

```python
ca_lines=[99.41,100.151,100.899,101.655,102.419,103.19,103.968,104.755,105.55,106.352,107.163,107.982,108.81,109.646,110.49,111.343,112.205,113.076,113.956,114.845,115.744,116.652,117.569,118.496,119.433,120.379,121.336,122.303,123.28,124.267,125.265,126.274,127.294,128.325]
```

```python
rms=np.nanstd(data['y'][300:400])
max_val=np.nanmax(data['y'][300:400])
min_val=np.nanmin(data['y'][300:400])
print rms, max_val, min_val
```

```python
bigfig=plt.figure(figsize=(10,5))
ax1=bigfig.add_subplot(111)
ax1.step(data['x'], data['y'],color='blue')
ax1.set_title("Blank Spectra",fontsize=20)
ax1.set_xlabel("Frequency (MHz)",fontsize=20)
ax1.set_xlim(103.2,104.2)
ax1.set_ylim(-1.5,1.5)
ax1.set_ylabel("Flux Density (Jy beam$^{-1}$)",fontsize=20)
ax1.tick_params(labelsize=20)
      
bigfig.savefig("Blank_Spectra.png")
```

```python

bigfig=plt.figure(figsize=(20,12))
ax1=bigfig.add_subplot(111)
ax1.step(data['x'], data['y'],color='blue')
ax1.fill_between(data['x'], 0, 1.5, where=data['x']==ca_lines[0], facecolor='green', alpha=0.5)
ax1.fill_between(data['x'], 0, 1.5, where=data['x']==ca_lines[1], facecolor='blue', alpha=0.5)
ax1.fill_between(data['x'], 0,1.5, where=data['x']==ca_lines[3], facecolor='red', alpha=0.5)
ax1.fill_between(data['x'], 0, 1.5, where=data['x']==ca_lines[4], facecolor='red', alpha=0.5)
ax1.set_title("",fontsize=18)
ax1.set_xlabel("Frequency (MHz)",fontsize=18)
ax1.set_xlim(99,125)
ax1.set_ylabel("Flux Density (Jy/beam)",fontsize=18)
ax1.tick_params(labelsize=15)
```
